# **Splitting the Dataset**

```python
X=data.drop(columns='target',axis=1)
Y=data['target']
```

```python
#printing the X and Y
print(X)
print(Y)
```

```python
#spliting the data in train test and split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, random_state=2, test_size=0.2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# **Model Train with Logistic Regression**

```python
model=LogisticRegression()
```

```python
#loading the LogisticRegression model training dataset
model.fit(X_train,Y_train)
```

# **Model Evaluation**

```python

X_train_prediction=model.predict(X_train)
```

```python
training_data_accuracy=accuracy_score(Y_train,X_train_prediction)
```

```python
#printing the accuracy score 
print('Accuracy score on training dataset: ',training_data_accuracy)
```

```python
#model Evealuation : accuracy score testing dataset
X_test_prediction=model.predict(X_test)
```

```python
test_data_accuracy=accuracy_score(Y_test,X_test_prediction)
```

```python
#printing the accuracy score 
print('Accuracy score on training dataset: ',test_data_accuracy)
```
